#  The Transaction Data Audit.

```python
from pathlib import Path
import polars as pl
import polars.selectors as cs
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from scipy import stats
import numpy as np

# ── Paths ─────────────────────────────────────────────────────────────────────
BASE         = Path("/home/brahimaskiou/Documents/jobsim/quantium/Data")
TXN_PATH     = BASE / "QVI_transaction_data.xlsx"
CUST_PATH    = BASE / "QVI_purchase_behaviour.csv"
OUTPUT_PATH  = BASE / "QVI_data.parquet"

# ══════════════════════════════════════════════════════════════════════════════
# TASK 1 — TRANSACTION DATA CLEANING
# ══════════════════════════════════════════════════════════════════════════════

# 1. Load
transactions = pl.read_excel(TXN_PATH)

# 2. Convert Excel integer date
transactions = transactions.with_columns(
    (pl.date(1899, 12, 30) + pl.duration(days=pl.col("DATE"))).alias("DATE")
)

# 3. Remove salsa
transactions = transactions.filter(
    ~pl.col("PROD_NAME").str.to_lowercase().str.contains("salsa")
)

# 4. Outlier check — 200 packets
outliers = transactions.filter(pl.col("PROD_QTY") == 200)
print("Commercial Outlier Transactions:")
print(outliers)

# Check what else that customer bought
suspect_card = outliers["LYLTY_CARD_NBR"][0]
print(f"\nAll transactions for card {suspect_card}:")
print(transactions.filter(pl.col("LYLTY_CARD_NBR") == suspect_card))

# Remove outlier
transactions = transactions.filter(pl.col("LYLTY_CARD_NBR") != suspect_card)

# 5. Pack size
transactions = transactions.with_columns(
    pl.col("PROD_NAME").str.extract(r"(\d+)", 1).cast(pl.Int32).alias("PACK_SIZE")
)

# 6. Brand
transactions = transactions.with_columns(
    pl.col("PROD_NAME").str.split(" ").list.get(0).alias("BRAND")
)

# 7. Clean brand names
brand_mapping = {
    "Red": "Red Rock Deli",
    "RRD": "Red Rock Deli",
    "Smith": "Smiths",
    "Snbts": "Sunbites",
    "Infzns": "Infuzions",
    "WW": "Woolworths",
    "Dorito": "Doritos",
    "Grain": "GrnWves",
}
transactions = transactions.with_columns(
    pl.col("BRAND").replace(brand_mapping)
)

print("\nBrand distribution:")
print(transactions.group_by("BRAND").agg(pl.len().alias("N")).sort("N", descending=True))

# ══════════════════════════════════════════════════════════════════════════════
# TRANSACTIONS OVER TIME — missing date detection
# ══════════════════════════════════════════════════════════════════════════════

transactions_by_day = (
    transactions.group_by("DATE")
    .agg(pl.len().alias("N"))
    .sort("DATE")
)
print(f"\nUnique dates in data: {transactions_by_day.height}")

# Build full date spine Jul 2018 → Jun 2019
full_dates = pl.DataFrame({
    "DATE": pl.date_range(
        start=pl.date(2018, 7, 1),
        end=pl.date(2019, 6, 30),
        interval="1d",
        eager=True
    )
})

transactions_by_day = full_dates.join(transactions_by_day, on="DATE", how="left").fill_null(0)

missing_dates = transactions_by_day.filter(pl.col("N") == 0)
print(f"Missing dates: {missing_dates['DATE'].to_list()}")

# Plot — full year
fig, ax = plt.subplots(figsize=(14, 4))
dates = transactions_by_day["DATE"].to_list()
counts = transactions_by_day["N"].to_list()
ax.plot(dates, counts, linewidth=0.8)
ax.set_title("Transactions Over Time")
ax.set_xlabel("Date")
ax.set_ylabel("Number of Transactions")
ax.xaxis.set_major_locator(mticker.MaxNLocator(12))
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig(BASE / "transactions_over_time.png", dpi=150)
plt.show()

# Plot — December zoom
dec_data = transactions_by_day.filter(
    (pl.col("DATE") >= pl.date(2018, 12, 1)) &
    (pl.col("DATE") <= pl.date(2018, 12, 31))
)
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(dec_data["DATE"].to_list(), dec_data["N"].to_list(), marker="o", linewidth=1)
ax.set_title("Transactions — December 2018")
ax.set_xlabel("Date")
ax.set_ylabel("Number of Transactions")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(BASE / "transactions_december.png", dpi=150)
plt.show()

# ══════════════════════════════════════════════════════════════════════════════
# PACK SIZE HISTOGRAM
# ══════════════════════════════════════════════════════════════════════════════

pack_dist = (
    transactions.group_by("PACK_SIZE")
    .agg(pl.len().alias("N"))
    .sort("PACK_SIZE")
)
fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(
    [str(x) for x in pack_dist["PACK_SIZE"].to_list()],
    pack_dist["N"].to_list(),
    color="steelblue"
)
ax.set_title("Transactions by Pack Size")
ax.set_xlabel("Pack Size (g)")
ax.set_ylabel("Number of Transactions")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(BASE / "pack_size_histogram.png", dpi=150)
plt.show()

# ══════════════════════════════════════════════════════════════════════════════
# CUSTOMER DATA
# ══════════════════════════════════════════════════════════════════════════════

customers = pl.read_csv(CUST_PATH)

print("\nCustomer schema:")
print(customers.schema)

# Null audit
null_counts = (
    customers.select([pl.all().is_null().sum()])
    .transpose(include_header=True, column_names=["null_count"])
)
print("\nNull audit:")
print(null_counts)

# Unique categories
print("\nUnique Lifestages:")
print(customers.select(pl.col("LIFESTAGE").unique()).sort("LIFESTAGE"))

print("\nUnique Premium statuses:")
print(customers.select(pl.col("PREMIUM_CUSTOMER").unique()).sort("PREMIUM_CUSTOMER"))

# Primary key check
total_rows  = customers.height
unique_cards = customers.select(pl.col("LYLTY_CARD_NBR").n_unique()).item()
print(f"\nTotal rows: {total_rows} | Unique cards: {unique_cards}")
print("✅ Unique PK" if total_rows == unique_cards else "⚠️ Duplicate cards found")

# Segment distribution
print("\nCustomer segment distribution:")
print(
    customers.group_by(["LIFESTAGE", "PREMIUM_CUSTOMER"])
    .agg(pl.len().alias("N_CUSTOMERS"))
    .sort(["LIFESTAGE", "PREMIUM_CUSTOMER"])
)

# ══════════════════════════════════════════════════════════════════════════════
# MERGE
# ══════════════════════════════════════════════════════════════════════════════

data = transactions.join(customers, on="LYLTY_CARD_NBR", how="left")

print(f"\nTransaction rows: {transactions.height} | Merged rows: {data.height}")
print("Nulls after merge:", data.select(pl.all().is_null().sum()).transpose(
    include_header=True, column_names=["null_count"]
).filter(pl.col("null_count") > 0))

# Save
data.write_parquet(OUTPUT_PATH)
print(f"\nSaved merged data to: {OUTPUT_PATH}")

# ══════════════════════════════════════════════════════════════════════════════
# ANALYSIS — SEGMENT INSIGHTS
# ══════════════════════════════════════════════════════════════════════════════

# Helper: grouped bar chart
def plot_segment(df, value_col, title, ylabel, filename):
    lifestages      = df["LIFESTAGE"].to_list()
    premium_vals    = df["PREMIUM_CUSTOMER"].to_list()
    values          = df[value_col].to_list()

    segments  = sorted(set(lifestages))
    premiums  = sorted(set(premium_vals))
    x         = np.arange(len(segments))
    width     = 0.25
    colors    = ["#2196F3", "#FF9800", "#4CAF50"]

    fig, ax = plt.subplots(figsize=(14, 5))
    for i, prem in enumerate(premiums):
        vals = []
        for seg in segments:
            row = df.filter(
                (pl.col("LIFESTAGE") == seg) &
                (pl.col("PREMIUM_CUSTOMER") == prem)
            )
            vals.append(row[value_col][0] if row.height > 0 else 0)
        ax.bar(x + i * width, vals, width, label=prem, color=colors[i])

    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.set_xticks(x + width)
    ax.set_xticklabels(segments, rotation=45, ha="right")
    ax.legend(title="Premium Status")
    plt.tight_layout()
    plt.savefig(BASE / filename, dpi=150)
    plt.show()


# 1. Total sales by segment
total_sales = (
    data.group_by(["LIFESTAGE", "PREMIUM_CUSTOMER"])
    .agg(pl.col("TOT_SALES").sum().alias("TOTAL_SALES"))
    .sort("TOTAL_SALES", descending=True)
)
print("\nTotal Sales by Segment:")
print(total_sales)
plot_segment(total_sales, "TOTAL_SALES", "Total Sales by Segment", "Total Sales ($)", "sales_by_segment.png")

# 2. Number of unique customers by segment
n_customers = (
    data.group_by(["LIFESTAGE", "PREMIUM_CUSTOMER"])
    .agg(pl.col("LYLTY_CARD_NBR").n_unique().alias("N_CUSTOMERS"))
    .sort("N_CUSTOMERS", descending=True)
)
print("\nCustomers by Segment:")
print(n_customers)
plot_segment(n_customers, "N_CUSTOMERS", "Number of Customers by Segment", "Unique Customers", "customers_by_segment.png")

# 3. Avg units per customer by segment
avg_units = (
    data.group_by(["LIFESTAGE", "PREMIUM_CUSTOMER"])
    .agg(
        (pl.col("PROD_QTY").sum() / pl.col("LYLTY_CARD_NBR").n_unique()).alias("AVG_UNITS")
    )
    .sort("AVG_UNITS", descending=True)
)
print("\nAvg Units per Customer by Segment:")
print(avg_units)
plot_segment(avg_units, "AVG_UNITS", "Avg Units per Customer by Segment", "Avg Units", "avg_units_by_segment.png")

# 4. Avg price per unit by segment
avg_price = (
    data.group_by(["LIFESTAGE", "PREMIUM_CUSTOMER"])
    .agg(
        (pl.col("TOT_SALES").sum() / pl.col("PROD_QTY").sum()).alias("AVG_PRICE")
    )
    .sort("AVG_PRICE", descending=True)
)
print("\nAvg Price per Unit by Segment:")
print(avg_price)
plot_segment(avg_price, "AVG_PRICE", "Avg Price per Unit by Segment", "Avg Price ($)", "avg_price_by_segment.png")

# ══════════════════════════════════════════════════════════════════════════════
# T-TEST — Mainstream vs Budget/Premium (young & midage singles/couples)
# ══════════════════════════════════════════════════════════════════════════════

target_lifestages = ["YOUNG SINGLES/COUPLES", "MIDAGE SINGLES/COUPLES"]

mainstream = (
    data.filter(
        (pl.col("PREMIUM_CUSTOMER") == "Mainstream") &
        (pl.col("LIFESTAGE").is_in(target_lifestages))
    )
    .with_columns((pl.col("TOT_SALES") / pl.col("PROD_QTY")).alias("PRICE_PER_UNIT"))
    ["PRICE_PER_UNIT"].to_numpy()
)

others = (
    data.filter(
        (pl.col("PREMIUM_CUSTOMER").is_in(["Budget", "Premium"])) &
        (pl.col("LIFESTAGE").is_in(target_lifestages))
    )
    .with_columns((pl.col("TOT_SALES") / pl.col("PROD_QTY")).alias("PRICE_PER_UNIT"))
    ["PRICE_PER_UNIT"].to_numpy()
)

t_stat, p_value = stats.ttest_ind(mainstream, others, equal_var=False)
print(f"\nT-test results:")
print(f"  t-statistic : {t_stat:.4f}")
print(f"  p-value     : {p_value:.6f}")
print(f"  Conclusion  : {'Significantly different ✅' if p_value < 0.05 else 'No significant difference ❌'}")

# ══════════════════════════════════════════════════════════════════════════════
# DEEP DIVE — Mainstream Young Singles/Couples
# ══════════════════════════════════════════════════════════════════════════════

target = data.filter(
    (pl.col("LIFESTAGE") == "YOUNG SINGLES/COUPLES") &
    (pl.col("PREMIUM_CUSTOMER") == "Mainstream")
)
rest = data.filter(
    ~(
        (pl.col("LIFESTAGE") == "YOUNG SINGLES/COUPLES") &
        (pl.col("PREMIUM_CUSTOMER") == "Mainstream")
    )
)

total_target = target.height
total_rest   = rest.height

# Brand affinity
def affinity(segment_df, population_df, col, seg_total, pop_total):
    seg_share = (
        segment_df.group_by(col)
        .agg((pl.len() / seg_total).alias("seg_share"))
    )
    pop_share = (
        population_df.group_by(col)
        .agg((pl.len() / pop_total).alias("pop_share"))
    )
    return (
        seg_share.join(pop_share, on=col)
        .with_columns((pl.col("seg_share") / pl.col("pop_share")).alias("affinity"))
        .sort("affinity", descending=True)
    )

brand_affinity = affinity(target, rest, "BRAND", total_target, total_rest)
print("\nBrand Affinity — Mainstream Young Singles/Couples:")
print(brand_affinity)

# Plot brand affinity
fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(
    brand_affinity["BRAND"].to_list()[::-1],
    brand_affinity["affinity"].to_list()[::-1],
    color="steelblue"
)
ax.axvline(1.0, color="red", linestyle="--", linewidth=1, label="Baseline (=1)")
ax.set_title("Brand Affinity — Mainstream Young Singles/Couples")
ax.set_xlabel("Affinity Score (>1 = over-indexed)")
ax.legend()
plt.tight_layout()
plt.savefig(BASE / "brand_affinity.png", dpi=150)
plt.show()

# Pack size affinity
pack_affinity = affinity(target, rest, "PACK_SIZE", total_target, total_rest)
print("\nPack Size Affinity — Mainstream Young Singles/Couples:")
print(pack_affinity)

# Plot pack size affinity
fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(
    [str(x) for x in pack_affinity["PACK_SIZE"].to_list()],
    pack_affinity["affinity"].to_list(),
    color="darkorange"
)
ax.axhline(1.0, color="red", linestyle="--", linewidth=1, label="Baseline (=1)")
ax.set_title("Pack Size Affinity — Mainstream Young Singles/Couples")
ax.set_xlabel("Pack Size (g)")
ax.set_ylabel("Affinity Score")
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(BASE / "pack_size_affinity.png", dpi=150)
plt.show()

print("\n✅ Task 1 complete — all outputs saved to:", BASE)
```
